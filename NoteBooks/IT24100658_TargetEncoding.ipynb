```python
df = pd.read_csv("IT24100603_HandleMissingData.csv")
```

```python
# Separate target variable
target = "Price"   # <-- replace with your actual target column
y = df[target]
X = df.drop(columns=[target])

# Apply Target Encoding
encoder = ce.TargetEncoder(cols=X.select_dtypes(include=['object']).columns)
X_encoded = encoder.fit_transform(X, y)

# Combine back target with encoded dataset
df_encoded = pd.concat([X_encoded, y], axis=1)

# Save
df_encoded.to_csv("IT24100658_TargetEncoding.csv", index=False)
print("Categorical variables encoded using Target Encoding. Saved as IT24100658_TargetEncoding.csv")
```

```python
corr = df_encoded.corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap After Target Encoding")
plt.show()
```
