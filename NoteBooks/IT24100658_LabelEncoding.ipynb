```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("IT24100603_HandleMissingData.csv")

label_enc = LabelEncoder()
for col in df.select_dtypes(include=['object']).columns:
    df[col] = label_enc.fit_transform(df[col].astype(str))

df.to_csv("IT24100658_LabelEncoding.csv", index=False)
print("Categorical variables encoded. Saved as IT24100658_LabelEncoding.csv")
```

```python
corr = df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap After Label Encoding")
plt.show()
```
