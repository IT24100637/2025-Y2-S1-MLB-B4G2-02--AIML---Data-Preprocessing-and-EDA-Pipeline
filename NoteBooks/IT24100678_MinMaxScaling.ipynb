```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("IT24100654_OutliersRemoving (1).csv")
```

```python
scaler = MinMaxScaler()

num_cols = df.select_dtypes(include=['int64', 'float64']).columns

df[num_cols] = scaler.fit_transform(df[num_cols])

df.to_csv("IT24100678_MinMaxScaling.csv", index=False)
print("Data scaled using MinMaxScaler. Saved as IT24100678_MinMaxScaling.csv")
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
target = "Price"

for col in num_cols:
    if col != target:
        plt.figure(figsize=(6, 4))
        sns.scatterplot(x=df[col], y=df[target], color="blue", alpha=0.6)
        plt.title(f"{col} vs {target} (After Scaling)")
        plt.xlabel(col)
        plt.ylabel(target)
        plt.show()
```
