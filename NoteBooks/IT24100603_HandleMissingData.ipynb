```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("MagicBricks.csv")
```

```python
print("Before removing duplicates:", df.shape)
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

```python
# Handling numeric values
df['Bathroom']=df['Bathroom'].fillna(df['Bathroom'].median())
df['Parking']=df['Parking'].fillna(df['Parking'].median())
df['Per_Sqft']=df['Per_Sqft'].fillna(df['Per_Sqft'].median())

# Handling categorical values
for col in ['Furnishing', 'Type']:
    df[col]=df[col].fillna(df[col].mode()[0])

df.to_csv("IT24100603_HandleMissingData.csv", index=False)
print("Missing values handled. Saved as IT24100603_HandleMissingData.csv")
```

```python
#Distribution of numerical features
num_cols = ['Area','BHK','Bathroom','Parking','Price','Per_Sqft']

plt.figure(figsize=(15,10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```
